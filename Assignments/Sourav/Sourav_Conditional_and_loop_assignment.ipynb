```python
#W. A P. which takes one number from 0 to 9 from the user and prints it in the word. And if the word is not from 0 to 9 then
#it should print that number is outside of the range and program should exit.
# For exapmple:- 
# input = 1
# output = one

user_input = int(input("Enter a number: "))
if user_input == 1:
    print("One")
elif user_input == 2:
    print("Two")
elif user_input == 3:
    print("Three")
elif user_input == 4:
    print("Four")
elif user_input == 5:
    print("Five")
elif user_input == 6:
    print("Six")
elif user_input == 7:
    print("Seven")
elif user_input == 8:
    print("Eight")
elif user_input == 9:
    print("Nine")
else:
    print("The number is outside of the range")
```

```python
#W. A P. to implement calculator but the operation to be done and two numbers will be taken as input from user:-
#Operation console should show below:-
#     Please select any one operation from below:-
#         * To add enter 1
#         * to subtract enter 2
#         * To multiply enter 3
#         * To divide enter 4
#         * To divide and find quotient enter 5
#         * To divide and find remainder enter 6
#         * To divide and find num1 to the power of num2 enter 7
#         * To Come out of the program enter 8


    
user_input1 = int(input("Enter a number 1: "))
user_input2 = int(input("Enter a number 2: "))
user_input3 = int(input('''Please select any one operation from below:-
         * To add enter 1
         * to subtract enter 2
         * To multiply enter 3
         * To divide enter 4
         * To divide and find quotient enter 5
         * To divide and find remainder enter 6
         * To divide and find num1 to the power of num2 enter 7
         * To Come out of the program enter 8

'''))
while(1):
  
    if user_input3 == 1:
        print(user_input1+user_input2)
        break
    elif user_input3 == 2:
        if user_input1>user_input2:
            print(user_input1-user_input2)
            break
        else:
            print(user_input2-user_input1)
            break
    elif user_input3 == 3:
        print(user_input1*user_input2)
        break
    elif user_input3 == 4:
        print(user_input1/user_input2)
        break
    elif user_input3 == 5:
        print(user_input1//user_input2)
        break
    elif user_input3 == 6:
        print(user_input1%user_input2)
        break
    elif user_input3 == 7:
        print(user_input1**user_input2)
        break
    elif user_input3 == 8:
        print("out of the program")
        break   
    else:
        print("Enter valid option")
        break


```

```python
#Write a Python program to find the length of the my_str using loop:-

#Input:- 'Write a Python program to find the length of the my_str'
#Output:- 55
input_str = 'Write a Python program to find the length of the my_str'
j = 0
for i in input_str:
    j=j+1
print("The length of the string : %s"%j)
```

```python
#Write a Python program to find the total number of times letter 'p' is appeared in the below string using loop:-
    
#Input:- 'peter piper picked a peck of pickled peppers.\n'
#Output:- 9

input_str = 'peter piper picked a peck of pickled peppers.\n'
j = 0
for i in input_str:
    if i == "p":
        j = j+1
print("The length of the string : %s"%j)
```

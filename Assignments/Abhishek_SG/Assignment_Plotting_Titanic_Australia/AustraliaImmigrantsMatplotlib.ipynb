```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\Admin\Desktop\dsc_weekday_1\Assignments\Abhishek_SG\Assignment_Plotting_Titanic_Australia\AustraliaFiltered.xlsx")
df
```

```python
# Plot bar chart of total no of immigrants for countires top 5 countries having max no of total immigrants
from textwrap import wrap
a = df.sort_values("Total", ascending=False).head()
b = a.groupby(by = "Country").groups.keys()
b = ['\n'.join(wrap(i,12)) for i in b]
# print(a)
plt.figure(figsize = (10,5))
plt.bar(a.Country,a.Total, color = ['g','orange', 'y', 'r', 'b'])
plt.xlabel("Country")
plt.xlabel("Immigrants")
plt.xticks(list(range(5)),b)
plt.show()
```

```python
# Find total immigrants from each category by area name and plot barchart for each area
from textwrap import wrap
a = df.groupby(by = "AreaName")['Total'].sum()

b = df.groupby(by = "AreaName").groups.keys()
b = ['\n'.join(wrap(i,12)) for i in b]

a.plot(kind ="bar", color = ['g','orange', 'y', 'r', 'b', 'm'])
plt.xticks([0,1,2,3,4,5],b)
plt.ylabel("Immigrants")
plt.show()
# df['cluster3'] = ['\n'.join(wrap(x, 12)) for x in  df['cluster3'])
```

```python
# Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot
plt.figure(figsize=(20,5))
a = df.loc[:,'1980':'2008'].sum()
a.plot(kind='line', marker = '.')
plt.xticks(list(range(0, len(a))), list(range(1980,2009)))
plt.show()
```

```python
# Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
a = df.groupby(by ="AreaName").sum().loc["Africa":"Europe", '1980':'2008']
b = a.unstack().unstack()

b.plot(kind='line', marker = '.',figsize=(20,5))
plt.xticks(list(range(0, len(b))), list(range(1980,2009)), rotation = 90)
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
# Plot Total Immigrants based on Development Region.
a = df.groupby(by = "Region")['Total'].sum()
plt.figure(figsize=(15,6))
a.plot(kind="bar")
plt.show()
```

```python
# Plot immigration trend for India using line plot
a = df.Country=='India'
b = df[a].loc[:,'1980':'2008'].sum()
plt.figure(figsize=(20,5))
b.plot(kind ='line', marker ='.')
plt.xticks(list(range(0, len(b))), list(range(1980,2009)))
plt.show()
```

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
a = df.sort_values("Total", ascending=False).head()
b = a.groupby(by="Country").sum().loc[:,'1980':'2008']
c = b.unstack().unstack()

c.plot(kind ="line", marker =".",figsize=(20,5))
plt.xticks(list(range(0, len(c))), list(range(1980,2009)), rotation = 90)
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")

plt.show()
```

```python
# For the above use area plot with alpha = 0.5


c.plot(kind ="area", alpha =0.5, stacked=False,figsize=(20,5))
plt.xticks(list(range(0, len(c))), list(range(1980,2009)), rotation = 90)
plt.show()
```

```python
# Compare Immigration Trend for top-5 countires using stacked area plot


c.plot(kind ="area", alpha =0.5, stacked=True, figsize=(20,5))
plt.xticks(list(range(0, len(c))), list(range(1980,2009)), rotation = 90)

plt.show()
```

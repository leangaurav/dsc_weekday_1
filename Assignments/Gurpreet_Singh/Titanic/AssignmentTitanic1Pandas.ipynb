```python
import pandas as pd
```

```python
# Read data and describe it
df= pd.read_csv(r"F:\AI_ML\dsc_weekday_1\Assignments\titanic_dataset.csv")
df.describe(include='all')
```

```python
# give me name of columns which are missing atleast one value
df.columns[df.isnull().any()]
```

```python
# print shape of dataset
print(df.shape)
```

```python
# group by gender and survived and see the counts in each category
print(df.groupby(["gender","survived"]).agg(['count']))

```

```python
# drop columns with more than 25% missing data
perc = len(df) * .75
var1= df.dropna(thresh = perc, axis = 1)
print(var1.columns)
print(df.columns)
print(var1['fare'].isnull().sum())
```

```python
# drop all columns with missing value
print(df.shape)
# delete all records that have at least  1 column data missing 
df.dropna(how='any',axis=0).shape
```

```python
#drop columns having independent values(which do not affect the survival rate).
# Assuming body , home.dest is something that do not affect the survival rate 
print(df.shape)
var1 = df.drop(columns=['body','home.dest'],axis=1)
print(var1.shape)
```

```python
#check data types of all columns
print(var1.dtypes)
```

```python
# Convert price to numeric  , price here is fare
df['fare']=df['fare'].str.replace('$','') # Treat as string and replace part of the string
df.fare.astype(float)
```

```python
#find columns still having missing/na values and also count of missing data
var1.isnull().sum()
```

```python
#fill na with mean for fare and age column column.
# r['fare'].value_counts(dropna=False)
# Nan below shows that 1 value is missing 
# print(var1['fare'].isnull().sum())
df['fare'].fillna(df['fare'].notnull().mean(), axis=0,inplace=True)
df['age'].fillna(df['age'].notnull().mean(), axis=0,inplace=True)
```

```python
print(df['age'].isnull().sum())
print(df['fare'].isnull().sum())
```

```python
df.to_csv('titanic_filtered.csv')
```

```python
# for survived column replace 0 with D and 1 with A
df['survived']  = df['survived'].replace(0,"D")
df[df['survived']=="D"]
df['survived']  = df['survived'].replace(1,"A")
df[df['survived']=="A"]
```

```python
df['survived'].value_counts()
```

```python
# group by gender and survived and see the counts in each category
df.groupby(['gender','survived']).size()
```

```python
#find different pclass and no of people in each class
df.groupby(['pclass']).size()
```

```python
#find top 5 people with highest values of age. Count no of male and females in the top 5
df.nlargest(5, 'age')
```

```python
x = df.nlargest(5, 'age')
x.groupby('gender').size()
```

```python
# find max age male and female who survived
df[["gender","age"]][df.survived =="A"].groupby('gender').max()
```

```python
#get average age by gender
df[["gender","age"]].groupby('gender').mean()
```

```python
# get average age by people survived vs not-survived
df[["survived","age"]].groupby('survived').mean()
```
